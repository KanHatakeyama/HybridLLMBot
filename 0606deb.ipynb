```python
db=SQLTextDB(initiate=False)
cross_encoder=CrossEncoder()
```

```python
import glob
text_path="UserData/original/text/kobunshi_ronbunshu/koron1974_47_12_47_12_1001.txt"
glob_path="UserData/original/text/kobunshi_ronbunshu/*.txt"
glob_path="UserData/original/test/*.txt"
for text_path in glob.glob(glob_path):
    db.add_text(text_path)
    print(text_path)
```

```python
query="高分子の特性は?"
query="藤田 正博先生の業績は?"
query="藤田 正博"
query="ポリフェニレンスルフィド"
#query="猫の名前は?"
query="ニャーと泣いていたのは誰?"
#query="藤田 正博先生の業績を纏めて下さい"
#query="藤田 正博先生, 業績"
#query="イオン液体"

print(db.tokenize_query(query))
hit_list=db.search_text(query)
ordered_answers=cross_encoder.predict(query,hit_list[:10])
```
