```python
import pandas as pd
```

```python
path="UserData/meta/qa/0617qa.txt"

with open(path, 'r') as f:
    text=f.read()

lines=text.split("\n")
```

```python
parsed_lines=[i.split("|||||") for i in lines]

df=pd.DataFrame(parsed_lines)
df.columns=["path","Q","A"]
df

for q,a in zip(df["Q"],df["A"]):
    print("---")
    print(q)
    print(a)
    print()
```

```python
#行をランダムに並び替える
df=df.sample(frac=1)
df[["Q","A"]]
```

```python
from sentence_transformers import SentenceTransformer, SentencesDataset, InputExample, losses, models
from sentence_transformers import SentenceTransformer, InputExample
from torch.utils.data import DataLoader

dataset=[(i,j) for i,j in zip(df["Q"],df["A"])]
train_examples = [InputExample(texts=[i[0],i[1]], label=1.0) for i in dataset]
train_dataloader = DataLoader(train_examples, shuffle=True, batch_size=16)
```

```python

from sentence_transformers import SentenceTransformer, util
q_id=2
def check_similarity(q_id,model):
    target_q=dataset[q_id][0]
    answers=[i[1] for i in dataset][:100]
    q_embeddings=model.encode(target_q)
    a_embeddings=model.encode(answers)
    cos_sim = util.cos_sim(q_embeddings, a_embeddings)[0]
    print(target_q)
    print(answers[q_id])
    a_df=pd.DataFrame(answers)
    a_df["cos_sim"]=cos_sim
    a_df.sort_values("cos_sim",ascending=False)
    return a_df
```

```python
bert = models.Transformer('sonoisa/sentence-bert-base-ja-mean-tokens-v2')
pooling = models.Pooling(bert.get_word_embedding_dimension())
model = SentenceTransformer(modules=[bert, pooling])
check_similarity(6,model)
```

```python
train_loss = losses.CosineSimilarityLoss(model)
#Tune the model
model.fit(train_objectives=[(train_dataloader, train_loss)], epochs=1, warmup_steps=100)
```

```python
check_similarity(6,model)
```
