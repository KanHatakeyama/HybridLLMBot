```python
import pandas as pd
from LLMSearch.Embedding.GPTEmbedding import GPTEmbedding
from settings.key import GPT_API_KEY
#https://zenodo.org/record/2783642
embedder=GPTEmbedding(GPT_API_KEY)
qa_df=pd.read_csv("develop/q_and_a_test.csv")

text_list=[]
text_list.extend(qa_df["Q"].tolist())
text_list.extend(qa_df["A"].tolist())
```

```python
import joblib
q_a_dict_path="develop/q_a_dict.bin"
text_dict={}
text_dict=joblib.load(q_a_dict_path)
for text in text_list:
    if text not in text_dict:
        print(text)
        vec=embedder(text)
        text_dict[text]=vec

        joblib.dump(text_dict,q_a_dict_path)
```
